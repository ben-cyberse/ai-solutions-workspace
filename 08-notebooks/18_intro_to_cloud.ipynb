## 2. What is Cloud Computing?

**NIST SP 800-145 Definition:**  
> *Cloud computing is a model for enabling ubiquitous, convenient, on-demand network access to a shared pool of configurable computing resources that can be rapidly provisioned and released with minimal management effort or service provider interaction.*

### Five Essential Characteristics (NIST)

| Characteristic | Description |
|----------------|-------------|
| **On-demand self-service** | Provision resources without human interaction from the provider |
| **Broad network access** | Available over the internet via standard devices |
| **Resource pooling** | Multi-tenant model — resources shared across customers |
| **Rapid elasticity** | Scale up/down automatically based on demand |
| **Measured service** | Pay-per-use metering — only pay for what you consume |

### Cloud vs On-Premises

| Factor | On-Premises | Cloud |
|--------|-------------|-------|
| **Capital cost** | High upfront (CapEx) | Low — operational expense (OpEx) |
| **Scaling** | Slow — buy hardware | Minutes — API call |
| **Maintenance** | Your team | Provider managed |
| **Availability** | Depends on your DR | Global regions, 99.99%+ SLAs |
| **Security control** | Full control | Shared responsibility |
| **Compliance** | Easier for strict data residency | Requires configuration |
| **Innovation speed** | Slower | Access to 200+ managed services |

```python
SERVICE_MODELS = {
    'IaaS': {
        'name'        : 'Infrastructure as a Service',
        'you_manage'  : ['OS', 'Runtime', 'Middleware', 'Applications', 'Data'],
        'provider_manages': ['Virtualisation', 'Servers', 'Storage', 'Networking'],
        'examples'    : ['AWS EC2', 'Azure Virtual Machines', 'GCP Compute Engine'],
        'best_for'    : 'Maximum control; lift-and-shift migrations'
    },
    'PaaS': {
        'name'        : 'Platform as a Service',
        'you_manage'  : ['Applications', 'Data'],
        'provider_manages': ['OS', 'Runtime', 'Middleware', 'Virtualisation', 'Servers', 'Storage', 'Networking'],
        'examples'    : ['AWS Elastic Beanstalk', 'Azure App Service', 'Google App Engine', 'Heroku'],
        'best_for'    : 'Developers who want to focus on code, not infrastructure'
    },
    'SaaS': {
        'name'        : 'Software as a Service',
        'you_manage'  : ['Data', 'User access configuration'],
        'provider_manages': ['Everything else'],
        'examples'    : ['Microsoft 365', 'Salesforce', 'Google Workspace', 'Slack', 'Zoom'],
        'best_for'    : 'End users and businesses wanting ready-to-use applications'
    },
    'FaaS': {
        'name'        : 'Function as a Service (Serverless)',
        'you_manage'  : ['Function code', 'Triggers'],
        'provider_manages': ['Servers', 'Scaling', 'OS', 'Runtime', 'Availability'],
        'examples'    : ['AWS Lambda', 'Azure Functions', 'Google Cloud Functions'],
        'best_for'    : 'Event-driven workloads; pay per execution (millisecond billing)'
    }
}

print('=== CLOUD SERVICE MODELS ===\n')
for model, info in SERVICE_MODELS.items():
    print(f'  [{model}] {info["name"]}')
    print(f'    You manage    : {info["you_manage"]}')
    print(f'    Provider manages: {info["provider_manages"][:3]}...')
    print(f'    Examples      : {", ".join(info["examples"][:3])}')
    print(f'    Best for      : {info["best_for"]}\n')
```

## 6. Core Cloud Concepts — Compute, Storage, Networking

```python
CLOUD_CONCEPTS = {
    'Compute': {
        'Virtual Machines'  : 'Full OS instance on shared hardware (EC2, Azure VM)',
        'Containers'        : 'Lightweight, portable process isolation (Docker, Kubernetes)',
        'Serverless'        : 'Code runs on demand; no server management (Lambda, Functions)',
        'Bare Metal'        : 'Dedicated physical hardware for performance/compliance workloads',
        'Auto Scaling'      : 'Automatically add/remove instances based on load metrics',
    },
    'Storage': {
        'Object Storage'    : 'Flat key-value store for unstructured data (S3, Blob) — infinite scale',
        'Block Storage'     : 'Disk volumes attached to VMs — high IOPS (EBS, Managed Disks)',
        'File Storage'      : 'Shared network filesystem (EFS, Azure Files)',
        'Archive Storage'   : 'Low-cost, infrequently accessed (S3 Glacier, Azure Archive)',
        'Data Redundancy'   : '3 copies minimum; replicated across AZs or regions',
    },
    'Networking': {
        'VPC / VNet'        : 'Isolated private network within the cloud (your own IP space)',
        'Subnets'           : 'Subdivisions of a VPC — public (internet-facing) vs private',
        'Security Groups'   : 'Stateful firewall rules at the instance level',
        'NACLs'             : 'Stateless firewall rules at the subnet level',
        'VPN / Direct Connect': 'Encrypted tunnel or dedicated line to on-premises network',
        'Peering'           : 'Connect two VPCs privately without internet transit',
        'CDN'               : 'Cache content at edge locations close to users (CloudFront)',
    }
}

for category, concepts in CLOUD_CONCEPTS.items():
    print(f'\n  {category.upper()}')
    print('  ' + '-'*55)
    for concept, definition in concepts.items():
        print(f'  {concept:<22}: {definition}')
```

## 8. Cloud Adoption Readiness Assessment

```python
# Simulated organisation cloud readiness assessment
READINESS_DIMENSIONS = [
    {'dimension': 'Business Readiness',     'score': 4, 'notes': 'Leadership buy-in secured; business case approved'},
    {'dimension': 'People & Skills',        'score': 2, 'notes': 'Only 2 staff with cloud certifications; training plan needed'},
    {'dimension': 'Governance & Compliance','score': 3, 'notes': 'Cloud policy drafted; compliance requirements mapped'},
    {'dimension': 'Security & Risk',        'score': 3, 'notes': 'Cloud security framework defined; controls partially implemented'},
    {'dimension': 'Technical Readiness',    'score': 3, 'notes': 'Applications assessed; 60% cloud-ready; 40% need refactoring'},
    {'dimension': 'Operational Readiness',  'score': 2, 'notes': 'No cloud monitoring or incident response process yet'},
    {'dimension': 'Financial Management',   'score': 2, 'notes': 'No FinOps process; cloud cost governance not established'},
]

total = sum(d['score'] for d in READINESS_DIMENSIONS)
max_total = len(READINESS_DIMENSIONS) * 5
pct = total / max_total * 100

print('=== CLOUD ADOPTION READINESS ASSESSMENT ===\n')
print(f'{"Dimension":<28} Score  Visual         Notes')
print('-' * 90)
for d in READINESS_DIMENSIONS:
    bar  = '█' * d['score'] + '░' * (5 - d['score'])
    flag = ' ⚠' if d['score'] <= 2 else ''
    print(f"  {d['dimension']:<26} {d['score']}/5  [{bar}]{flag}")
    print(f"    {d['notes']}")

level = 'READY' if pct >= 75 else 'PARTIALLY READY' if pct >= 55 else 'NOT READY'
print(f'\nOverall Score: {total}/{max_total} ({pct:.0f}%)  —  {level}')
print('Priority gaps: People & Skills, Operational Readiness, Financial Management')
```
